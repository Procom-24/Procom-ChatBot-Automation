```python
intentss = json.loads(open('data.json').read()) 
import os
import openai
import random
import json
from dotenv import load_dotenv, find_dotenv

_ = load_dotenv(find_dotenv())

openai.api_key = '' 

def get_completion(prompt, model="gpt-3.5-turbo"):
    messages = [{"role": "user", "content": prompt}]
    response = openai.ChatCompletion.create(
        model=model,
        messages=messages,
        temperature=0  # the degree of randomness of the model's output
    )
    return response.choices[0].message["content"]

def get_response(pattern, intents):
    for intent in intents['intents']:
        if pattern in intent['patterns']:
            return random.choice(intent['responses'])

    return "I apologize, but I couldn't comprehend your input."

# Example usage

pattern = "Tell me a Joke"
prompt = f"""
Your task is to help a user, who will provide one of the patterns,
the pattern is {pattern}, now match it with the patterns given in the intents inside the intents of {intentss}, use the function {get_response(pattern, intentss)} 
to get the response and modify it a bit more with your intelligence to make it sound more professional.

Now, assume I am the user, and you are the chatbot.
I have asked you a question which is {pattern},
now you will provide the response.
"""

response = get_completion(prompt)
print(response)
 
```
